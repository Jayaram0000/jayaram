```python
# Define a function to detect outliers using the IQR method
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column] <= lower_bound) & (data[column] >= upper_bound)]
    return outliers

# Get numerical columns for outlier analysis
numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns

# Initialize a figure for visualizing the outliers
plt.figure(figsize=(15, len(numerical_columns) * 5))

# Loop through each numerical column and create boxplots
outliers_summary = {}
for i, column in enumerate(numerical_columns, 1):
    plt.subplot(len(numerical_columns), 1, i)
    sns.boxplot(x=df[column], color="skyblue", orient='h')
    plt.title(f"Boxplot for {column}")

    # Detect and store outliers for the column
    outliers_summary[column] = detect_outliers_iqr(df, column)[column].values

plt.tight_layout()
plt.show()
```

```python
Q1 = df['Copper'].quantile(0.25)
Q3 = df['Copper'].quantile(0.75)
IQR = Q3 - Q1

# Determine the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Find the median of the Copper column
copper_median = df['Copper'].median()

# Replace outliers with the median
df['Copper'] = df['Copper'].apply(
    lambda x: copper_median if x < lower_bound or x > upper_bound else x
)
```

```python
q1=df['Platelets_log'].quantile(0.25)
q3=df['Platelets_log'].quantile(0.75)
iqr=q3-q1
lower_bound=q1-1.5*iqr
upper_bound=q3+1.5*iqr
print(lower_bound,upper_bound)
df2=df
df2['Platelets_log']=np.where(df2['Platelets_log']>upper_bound,upper_bound,df2['Platelets_log'])
df2['Platelets_log']=np.where(df2['Platelets_log']<lower_bound,lower_bound,df2['Platelets_log'])
print("shape after removing outliers",df2.shape)
sns.boxplot(df2['Platelets_log'])
plt.show()
```

```python
# Calculate the IQR for the Tryglicerides column
Q1 = df['Tryglicerides'].quantile(0.25)
Q3 = df['Tryglicerides'].quantile(0.75)
lower_bound = Q1
upper_bound = Q3

# Clip the values to the quartile range
df['Tryglicerides'] = df['Tryglicerides'].clip(lower=lower_bound, upper=upper_bound)
sns.boxplot(df['Tryglicerides'])
```

```python
columns = ['Cholesterol_log', 'Albumin_log', 'Copper_log', 'Alk_Phos_log', 'SGOT_log','Prothrombin_log','Tryglicerides_log']

# Apply clipping for each column based on its own Q1 and Q3
for col in columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    lower_bound = Q1
    upper_bound = Q3

    # Clip the column values to the quartile range
    df[col] = df[col].clip(lower=lower_bound, upper=upper_bound)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
# Define features (X) and target (y)
X = df.drop(columns=['Status'])  # Exclude the target column
y = df['Status']  # Target column: Stage

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)

# Initialize and train
rf_c = RandomForestClassifier(n_estimators=1000,         # Number of trees
    max_depth=20,             # Maximum depth of each tree
    min_samples_split=10,     # Minimum samples to split a node
    min_samples_leaf=5,       # Minimum samples required in a leaf
    max_features='sqrt',      # Number of features to consider for best split
    bootstrap=True,           # Use bootstrap sampling
    oob_score=True,           # Enable out-of-bag score for validation
    random_state=42)
rf_c.fit(X_train, y_train)

# Make predictions
y_pred = rf_c.predict(X_test)
from sklearn.metrics import accuracy_score, classification_report

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.metrics import accuracy_score, classification_report
# Define StratifiedKFold
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Evaluate the model using cross_val_score
scores = cross_val_score(rf_c, X, y, cv=skf, scoring='accuracy')

# Print the mean accuracy and standard deviation
print(f"Cross-Validation Accuracy: {scores.mean():.2f} ± {scores.std():.2f}")

# Train and evaluate on the full dataset (optional, for final model)
rf_c.fit(X_train, y_train)
y_pred = rf_c.predict(X_test)
print(f"Final Model Accuracy: {accuracy_score(y_test, y_pred):.2f}")
print(classification_report(y_test, y_pred))
```

```python
y_train_pred = rf_c.predict(X_train)  # Predictions on training data
train_accuracy = accuracy_score(y_train, y_train_pred) * 100  # Convert to percentage

# Testing accuracy
y_test_pred = rf_c.predict(X_test)  # Predictions on testing data
test_accuracy = accuracy_score(y_test, y_test_pred) * 100  # Convert to percentage

print(f"Training Accuracy: {train_accuracy:.2f}%")
print(f"Testing Accuracy: {test_accuracy:.2f}%")
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import StratifiedKFold, GridSearchCV
from sklearn.metrics import accuracy_score, classification_report

# Define parameter grid for GridSearchCV
param_grid = {
    'C': [0.1, 1, 10],               # Regularization parameter
    'kernel': ['linear', 'rbf'],     # Kernel type
    'gamma': ['scale', 'auto']       # Kernel coefficient (used for 'rbf')
}

# Initialize Stratified K-Fold
kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Lists to store accuracy scores
train_accuracies = []
test_accuracies = []

# Perform Stratified K-Fold Cross-Validation
for fold, (train_idx, test_idx) in enumerate(kf.split(X, y), 1):
    # Split the data into training and validation sets for the current fold
    X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
    y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]

    # Initialize GridSearchCV
    grid_search = GridSearchCV(
        estimator=SVC(),
        param_grid=param_grid,
        scoring='accuracy',
        cv=3,              # Inner cross-validation (3 folds)
        n_jobs=-1,         # Use all available cores
        verbose=1          # Print progress
    )

    # Fit GridSearchCV to the training data
    grid_search.fit(X_train, y_train)

    # Get the best model
    best_model = grid_search.best_estimator_

    # Make predictions on training and validation sets
    y_train_pred = best_model.predict(X_train)
    y_test_pred = best_model.predict(X_test)

    # Calculate accuracies
    train_accuracy = accuracy_score(y_train, y_train_pred)
    test_accuracy = accuracy_score(y_test, y_test_pred)

    # Store the accuracies
    train_accuracies.append(train_accuracy)
    test_accuracies.append(test_accuracy)

    # Print fold results
    print(f"Fold {fold}:")
    print(f"  Best Parameters: {grid_search.best_params_}")
    print(f"  Training Accuracy: {train_accuracy:.2f}")
    print(f"  Test Accuracy: {test_accuracy:.2f}")
    print("  Classification Report:")
    print(classification_report(y_test, y_test_pred))
    print("-" * 50)

# Average accuracies across folds
avg_train_accuracy = sum(train_accuracies) / len(train_accuracies)
avg_test_accuracy = sum(test_accuracies) / len(test_accuracies)

# Print overall results
print("\nOverall Results:")
print(f"Average Training Accuracy: {avg_train_accuracy:.2f}")
print(f"Average Test Accuracy: {avg_test_accuracy:.2f}")
```
