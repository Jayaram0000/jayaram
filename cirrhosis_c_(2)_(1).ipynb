```python
# Define a function to detect outliers using the IQR method
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column] <= lower_bound) & (data[column] >= upper_bound)]
    return outliers

# Get numerical columns for outlier analysis
numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns

# Initialize a figure for visualizing the outliers
plt.figure(figsize=(15, len(numerical_columns) * 5))

# Loop through each numerical column and create boxplots
outliers_summary = {}
for i, column in enumerate(numerical_columns, 1):
    plt.subplot(len(numerical_columns), 1, i)
    sns.boxplot(x=df[column], color="skyblue", orient='h')
    plt.title(f"Boxplot for {column}")

    # Detect and store outliers for the column
    outliers_summary[column] = detect_outliers_iqr(df, column)[column].values

plt.tight_layout()
plt.show()
```

**Decision Tree**

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split, StratifiedKFold, GridSearchCV
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_curve, auc
from sklearn.preprocessing import label_binarize
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Split the dataset into features (X) and target (y)
# Assuming 'Status' is the target column
X = df.drop(columns=['Status'])  # Replace 'df' with your actual DataFrame
y = df['Status']

# Split the data into training and testing sets (for final evaluation)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)

# Define StratifiedKFold for consistent distribution of classes
skf = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)

# Initialize Decision Tree model
dt_model = DecisionTreeClassifier(random_state=42)

# Define parameter grid for GridSearchCV
param_grid = {
    'max_depth': [5, 10, 15, 20, None],  # Tree depth
    'min_samples_split': [2, 5, 10],    # Minimum number of samples required to split a node
    'min_samples_leaf': [1, 2, 5],      # Minimum number of samples required to be at a leaf node
    'criterion': ['gini', 'entropy']    # Split quality criterion
}

# Set up GridSearchCV with StratifiedKFold for cross-validation
grid_search = GridSearchCV(
    estimator=dt_model,
    param_grid=param_grid,
    cv=skf,
    scoring='accuracy',
    n_jobs=-1,        # Use all CPU cores for faster execution
    verbose=1         # Print grid search progress
)

# Fit the model using GridSearchCV
grid_search.fit(X_train, y_train)

# Get the best parameters and model
best_params = grid_search.best_params_
best_model = grid_search.best_estimator_

# Evaluate the best model on the training set
y_train_pred = best_model.predict(X_train)
train_accuracy = accuracy_score(y_train, y_train_pred) * 100  # Convert to percentage

# Evaluate the best model on the testing set
y_test_pred = best_model.predict(X_test)
test_accuracy = accuracy_score(y_test, y_test_pred) * 100  # Convert to percentage

# Calculate overall accuracy (combine train and test predictions)
X_combined = pd.concat([X_train, X_test])  # Combine training and testing features
y_combined = pd.concat([y_train, y_test])  # Combine training and testing target
y_combined_pred = best_model.predict(X_combined)  # Predictions for combined data
overall_accuracy = accuracy_score(y_combined, y_combined_pred) * 100  # Convert to percentage

# Print results
print(f"Best Hyperparameters: {best_params}")
print(f"Training Accuracy: {train_accuracy:.2f}%")
print(f"Testing Accuracy: {test_accuracy:.2f}%")
print(f"Overall Accuracy: {overall_accuracy:.2f}%")

# Print Classification Report for Testing Data
print("\nClassification Report (Test Data):")
print(classification_report(y_test, y_test_pred))

# 1. Confusion Matrix
cm = confusion_matrix(y_test, y_test_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=y.unique(), yticklabels=y.unique())
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()

# 2. ROC Curve and AUC (multi-class)
# Binarize the labels for multi-class classification
y_test_bin = label_binarize(y_test, classes=y.unique())  # Binarize the labels for multi-class
y_score = best_model.predict_proba(X_test)

# ROC Curve for each class
fpr, tpr, _ = roc_curve(y_test_bin.ravel(), y_score.ravel())
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', lw=2, label=f'ROC curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='gray', lw=2, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc='lower right')
plt.show()

```

**Random Forest**

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_curve, auc
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from termcolor import colored
from sklearn.preprocessing import label_binarize

# Define features (X) and target (y)
X = df.drop(columns=['Status'])  # Exclude the target column
y = df['Status']  # Target column: Status

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)

# Initialize and train RandomForestClassifier
rf_c = RandomForestClassifier(
    n_estimators=1000,         # Number of trees
    max_depth=20,              # Maximum depth of each tree
    min_samples_split=10,      # Minimum samples to split a node
    min_samples_leaf=5,        # Minimum samples required in a leaf
    max_features='sqrt',       # Number of features to consider for best split
    bootstrap=True,            # Use bootstrap sampling
    oob_score=True,            # Enable out-of-bag score for validation
    random_state=42
)

# Perform cross-validation (using 5 folds by default)
cv_scores = cross_val_score(rf_c, X_train, y_train, cv=5, scoring='accuracy')

# Train the model on the full training data
rf_c.fit(X_train, y_train)

# Make predictions
y_train_pred = rf_c.predict(X_train)
y_test_pred = rf_c.predict(X_test)

# Calculate accuracies
train_accuracy = accuracy_score(y_train, y_train_pred) * 100  # Convert to percentage
test_accuracy = accuracy_score(y_test, y_test_pred) * 100      # Convert to percentage

# Calculate overall accuracy (combining train and test predictions)
X_combined = pd.concat([X_train, X_test])  # Combine training and testing features
y_combined = pd.concat([y_train, y_test])  # Combine training and testing target
y_combined_pred = rf_c.predict(X_combined)  # Predictions for combined data
overall_accuracy = accuracy_score(y_combined, y_combined_pred) * 100  # Convert to percentage

# Print results with colored output
print(colored(f"Cross-Validation Accuracy: {cv_scores.mean():.2f} ± {cv_scores.std():.2f}", 'cyan'))
print(colored(f"Training Accuracy: {train_accuracy:.2f}%", 'green'))
print(colored(f"Testing Accuracy: {test_accuracy:.2f}%", 'yellow'))
print(colored(f"Overall Accuracy: {overall_accuracy:.2f}%", 'magenta'))

# Print Classification Report for Testing Data
print(colored("\nClassification Report (Test Data):", 'blue'))
classification_report_str = classification_report(y_test, y_test_pred)
# Color classification report output
print(colored(classification_report_str, 'white'))

# Confusion Matrix
cm = confusion_matrix(y_test, y_test_pred)
sns.heatmap(cm, annot=True, fmt="d", cmap='Blues', cbar=False)
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

# ROC Curve and AUC
# For multi-class classification, we need to binarize the labels
y_test_bin = label_binarize(y_test, classes=y.unique())
y_pred_prob = rf_c.predict_proba(X_test)

# Compute ROC curve and AUC for each class
fpr, tpr, roc_auc = {}, {}, {}
for i in range(y_test_bin.shape[1]):
    fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_pred_prob[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# Plot the ROC curve
plt.figure(figsize=(10, 6))
for i in range(y_test_bin.shape[1]):
    plt.plot(fpr[i], tpr[i], color=plt.cm.jet(i / y_test_bin.shape[1]), lw=2, label=f'Class {i} (AUC = {roc_auc[i]:.2f})')

plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()

```

**Naive Bayes Classifier**

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import (
    accuracy_score, classification_report, confusion_matrix, roc_curve, auc
)
from sklearn.model_selection import StratifiedKFold, GridSearchCV
from sklearn.preprocessing import label_binarize
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Split the dataset into features (X) and target (y)
X = df.drop('Status', axis=1)  # Replace 'df' with your actual DataFrame
y = df['Status']  # Target column: Status

# Initialize the Gaussian Naive Bayes model
nb_model = GaussianNB(priors=None)  # No prior probabilities

# Define the hyperparameter grid for GridSearchCV
param_grid = {
    'var_smoothing': [1e-12, 1e-9, 1e-6, 1e-3, 1e-1]  # Try different values for var_smoothing
}

# Initialize StratifiedKFold
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Initialize GridSearchCV
grid_search = GridSearchCV(estimator=nb_model, param_grid=param_grid, cv=skf, scoring='accuracy', n_jobs=-1, verbose=1)

# Lists to store metrics
train_accuracies, test_accuracies = [], []
final_y_train, final_y_train_pred = [], []
final_y_test, final_y_test_pred = [],[]
final_y_test_proba = []

# Perform Grid Search with Stratified Cross-Validation
fold = 1
for train_index, test_index in skf.split(X, y):
    # Split data into training and testing sets for this fold
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    # Fit GridSearchCV on the training data
    grid_search.fit(X_train, y_train)

    # Get the best model
    best_model = grid_search.best_estimator_

    # Training Accuracy
    y_train_pred = best_model.predict(X_train)
    train_accuracy = accuracy_score(y_train, y_train_pred) * 100
    train_accuracies.append(train_accuracy)

    # Testing Accuracy
    y_test_pred = best_model.predict(X_test)
    test_accuracy = accuracy_score(y_test, y_test_pred) * 100
    test_accuracies.append(test_accuracy)

    # Probabilities for ROC-AUC Curve
    y_test_proba = best_model.predict_proba(X_test)

    # Store results for overall evaluation
    final_y_train.extend(y_train)
    final_y_train_pred.extend(y_train_pred)
    final_y_test.extend(y_test)
    final_y_test_pred.extend(y_test_pred)
    final_y_test_proba.append(y_test_proba)

    # Print fold results
    print(f"Fold {fold}:")
    print(f"Best Hyperparameters: {grid_search.best_params_}")
    print(f"Training Accuracy: {train_accuracy:.2f}%")
    print(f"Testing Accuracy: {test_accuracy:.2f}%")
    print(classification_report(y_test, y_test_pred))
    fold += 1

# Combine predicted probabilities
final_y_test_proba = np.vstack(final_y_test_proba)

# Overall Metrics
final_train_accuracy = accuracy_score(final_y_train, final_y_train_pred) * 100
final_test_accuracy = accuracy_score(final_y_test, final_y_test_pred) * 100
overall_accuracy = accuracy_score(final_y_train + final_y_test, final_y_train_pred + final_y_test_pred) * 100

print("\nOverall Results:")
print(f"Final Training Accuracy: {final_train_accuracy:.2f}%")
print(f"Final Testing Accuracy: {final_test_accuracy:.2f}%")
print(f"Overall Accuracy: {overall_accuracy:.2f}%")

# Confusion Matrix
cm = confusion_matrix(final_y_test, final_y_test_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=np.unique(y), yticklabels=np.unique(y))
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix")
plt.show()

# Binarize test labels for multi-class ROC-AUC
y_bin = label_binarize(final_y_test, classes=np.unique(y))
n_classes = y_bin.shape[1]

# Compute ROC curve and AUC for each class
fpr, tpr, roc_auc = dict(), dict(), dict()
for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_bin[:, i], final_y_test_proba[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# Plot ROC curve for each class
plt.figure(figsize=(10, 8))
colors = ['blue', 'green', 'red', 'orange', 'purple']  # Adjust based on class count
for i in range(n_classes):
    plt.plot(fpr[i], tpr[i], color=colors[i % len(colors)], lw=2,
             label=f'Class {np.unique(y)[i]} (AUC = {roc_auc[i]:.2f})')

# Random Guess Line
plt.plot([0, 1], [0, 1], color='gray', lw=2, linestyle='--')

# Customize Plot
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Multi-Class ROC-AUC Curve")
plt.legend(loc="lower right")
plt.show()


```
